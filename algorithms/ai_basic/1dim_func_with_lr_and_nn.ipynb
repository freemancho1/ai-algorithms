```python
import tensorflow as tf
import numpy as np

from sklearn.linear_model import LinearRegression

tf.__version__
```

```python
train_x = np.array([[5], [-10], [9], [0], [4], [2]])
train_y = np.array([17, -28, 29, 2, 14, 8])   # train_y = train_x * 3 + 2
train_x.shape, train_x, train_y, train_y.shape
```

## Linear Regression

```python
lr = LinearRegression()
lr.fit(train_x[:2], train_y[:2])
```

```python
pred_lr = lr.predict(np.array([[1], [3]]))
pred_lr
```

## NN

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=1, input_shape=(1,)))
model.compile(optimizer="sgd", loss="mse", metrics=["accuracy"])
model.summary()
```

```python
model.fit(train_x, train_y, epochs=200)
```

```python
pred_y = model.predict(np.array([[1],[3]]))
```
